```python
#Notation -- EllipticCurve([a1,a2,a3,a4,a6]) -- y^2 + a1xy + a3y = x^3 + a2x^2 + a4x + a6
#E = EllipticCurve(CC,j=0);E
E = EllipticCurve([0,0,0,0,1]);E
```

```python
E.plot(thickness=1, rgbcolor=(1,0,0))
```

```python
P = E(0,1);P
```

```python
E.integral_points()
```

```python
Q = E(-1,0);Q
```

```python
P+Q
```

```python
g = Graphics()
#Plot Curve
g += E.plot(thickness=1, rgbcolor=(1,0,0))

#Plot Points
g += P.plot(size = 200, rgbcolor=(0,1,0))
g += Q.plot(size = 200, rgbcolor=(0,0,1))

#Plot Sum
g += (P+Q).plot(size = 200, rgbcolor=(0,1,1)) #Teal

#Plot Lines Between
g += line2d([(Q.xy()[0],Q.xy()[1]),((-(P+Q)).xy()[0],(-(P+Q)).xy()[1])],thickness = 2, rgbcolor=(3,2,1))
g += line2d([((P+Q).xy()[0],(P+Q).xy()[1]),((-(P+Q)).xy()[0],(-(P+Q)).xy()[1])],thickness = 2, rgbcolor=(3,2,1))
#Show
g
```
